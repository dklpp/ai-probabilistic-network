```python
from mrmr import mrmr_classif
import pandas as pd

list = ["activity_train.csv"]
#"example2.csv", "example3.csv",

for i in list:
    df = pd.read_csv(i)
    print(df.shape)
    print(df.head())

    X = df.iloc[:, 0:-1]
    y = df.iloc[:, -1]

    number_of_selected_features = 100
    selected_features = mrmr_classif(X=X, y=y, K=number_of_selected_features)
    selected_features.append('activity')
    print(selected_features)

    df = df[selected_features]
    df.to_csv('MRMR_selected_'+str(number_of_selected_features)+'_'+str(i))
```
